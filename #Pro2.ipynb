```python
import numpy as np
import matplotlib.pyplot as plt
```

1 Numerical derivative 
2 Stiff kinetics

```python
dt = 0.01

def E_m(f,f_1d):
    return f + dt*f_1d

t = np.arange(0,0.5,0.01)
# print(t)

f = np.zeros((3,50))
# print(f)
f[:,0] = np.array([0.6,0.1,0.3])

def x(t):
    return np.exp(-t) - 0.4

def y_1d(x,y,z):
    return x - 5*y + 10*z

def z_1d(y,z):
    return 5*y - 10*z

f[0] = x(t)
# print(f[0])

for i in range(0,49):
    f[1][i+1] = E_m(f[1][i],y_1d(f[0][i], f[1][i], f[2][i]))
    f[2][i+1] = E_m(f[2][i],z_1d(f[1][i], f[2][i]))

# print(f)
```

```python
plt.plot(t,f[1], label = "y(t)", alpha=0.8)
plt.plot(t,f[2], label = "z(t)", alpha=0.8)
plt.scatter(t,f[1], s=8, alpha=0.8)
plt.scatter(t, f[2], s=8, alpha=0.8)
plt.xlabel("t")
plt.legend()

plt.show()
```
